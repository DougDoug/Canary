## Recording

```python
from IPython.display import Image
Image("../rapport/Images/Creation Et Verif Data/0_Les_donnees.png")
```

Here's the data, each line correspond to a song. The view given here is as a spectrogram. This view is convenient because it gives us every informations we want to reach the *atomos* of the song and to identify its composition. The colors you can see inform you about the power of the signal, if it's loud or quiet. The x axis is the time *per se* and the y axis is the **pitch**. The higher the points are on this axis, the higher their frequency are.

Let's observe the last song, it begins with 5 horizontal sticks. Each stick is a syllable. It's separated with small spaces and are repetitives. If we don't listen we might think that this group of five stick forms a sentence. But actually... *It's a trap !*. If we listen to this song we'll easily hear that the two seconds sticks are marked by a sudden changing in pitch. We can see the pitch variation but it's quite easier to hear it (and avoid a lot of mistakes).

## Labels

```python
Image("../rapport/Images/Creation Et Verif Data/1 Mettre des labels.png")
```

We now zoom dramatically into every song and put some labels. A label is built on the **index of the song** it belongs and its **sentence type**. Namely we have selected here a syllable on the song number 19, and this syllable belong to a sentence of the type "F10". We often put some digits after the rigourous name of the syllable it means that this syllable possess more or less sudden variations without making a break. It means that by hear we can differ a "F10" of a "F1" but also that visualy they often share a common "shape".
